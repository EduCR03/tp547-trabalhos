## Questao 1 - Gerador Linear Congruente

Considere um gerador linear congruente definido por `X_(n+1) = (aX_n + c) mod m`, com `a = 11`, `c = 7`, `m = 32` e `X0 = 5`.

### a)

Calcule os 10 primeiros numeros gerados pela sequencia.


```python
import numpy as np
import matplotlib.pyplot as plt

x=5
x1=np.array([x])
n=32
a=11
c=7
m=32

for i in range(n):
    x=(a*x+c)%m
    x1=np.append(x1,x)

print(x1)
print("Os 10 primeiros numeros gerados sao")
print(x1[1:11])

ind=np.arange(n+1)
plt.bar(ind, x1)
plt.show()
```

### b) e c)

Determine o periodo do gerador e verifique se o gerador satisfaz o periodo maximo.


```python
periodo=0
for i in range(1,len(x1)):
    if x1[i]==x1[0]:
        periodo = i
        break

print("O periodo do gerador e", periodo)

if periodo==m:
    print("O gerador satisfaz o periodo maximo.")
else:
    print("O gerador nao satisfaz o periodo maximo.")
```

### d)

Discuta limitacoes desse gerador para simulacoes de Monte Carlo em larga escala.


```python
valores=np.arange(m)
freq=np.array([])

for valor in valores:
    cont=0
    for ix in x1[0:periodo]:
        if ix==valor:
            cont=cont+1
    freq=np.append(freq,cont)

print("Frequencia dos valores em um periodo")
print(freq)

plt.bar(valores, freq)
plt.show()

print("Limitacoes para simulacoes de Monte Carlo em larga escala:")
print("O periodo maximo seria", m, "e o periodo encontrado foi", periodo)
print("A sequencia repete rapido para simulacoes longas e o gerador nao visita todos os valores de 0 ate", m-1)
```
