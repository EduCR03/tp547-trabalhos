### a)

Qual e a probabilidade de ocorrerem exatamente 6 falhas em uma hora?


```python
import numpy as np
import matplotlib.pyplot as plt

lambda1=4 #Numero medio de falhas
N=100000 #Numero de amostras
value=6
count=0
av=np.array([])
x=np.random.uniform(0,1,N)
for ix in x:
    i = 0
    pr = np.exp(-lambda1)
    F=pr
    while ix>=F:
        pr=lambda1/(i+1)*pr
        F = F + pr
        i = i + 1;
    a1=i
    av=np.append(av,a1)

print(av)

for expvalue in av:
    if expvalue == value:
            count=count+1
prob=count/N

print("a probabilidade de ocorrerem exatamente 6 falhas em uma hora e", prob)

valores=np.arange(0,16)
freq=np.array([])
for valor in valores:
    count2=0
    for expvalue in av:
        if expvalue == valor:
            count2=count2+1
    freq=np.append(freq,count2/N)

plt.bar(valores, freq)
plt.show()
```
