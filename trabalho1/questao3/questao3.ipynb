### Simulacao Base

Gerar amostras da distribuicao binomial para o caso de 20 pacotes com probabilidade de erro igual a 0.08.


```python
import numpy as np
import matplotlib.pyplot as plt

q=0.08#Probabilidade de erro
n=20 #Numero de pacotes
N=100000#Numero de amostras
c = q/(1-q)
av=np.array([])
x=np.random.uniform(0,1,N)
for ix in x:
    i = 0
    pr = pow((1 - q),n)
    F = pr
    while ix>=F:
        pr = (c * (n - i) / (i + 1))* pr;
        F = F + pr;
        i = i + 1;
    a1=i
    av=np.append(av,a1)

print(av)

valores=np.arange(0,n+1)
freq=np.array([])
for valor in valores:
    count=0
    for binvalue in av:
        if binvalue==valor:
            count=count+1
    freq=np.append(freq,count/N)

plt.bar(valores, freq)
plt.show()
```

### d)

Calcule media e variancia da distribuicao.


```python
q=0.08#Probabilidade de erro
n=20 #Numero de pacotes
N=100000#Numero de amostras
c = q/(1-q)
av=np.array([])
x=np.random.uniform(0,1,N)
for ix in x:
    i = 0
    pr = pow((1 - q),n)
    F = pr
    while ix>=F:
        pr = (c * (n - i) / (i + 1))* pr;
        F = F + pr;
        i = i + 1;
    a1=i
    av=np.append(av,a1)

print("a media teorica e",n*q)
print("a variancia teorica e",n*q*(1-q))
print("a media amostral e",np.mean(av))
print("a variancia amostral e",np.var(av))
```
