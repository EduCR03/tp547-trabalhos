### a)

Criar uma simulacao para gerar amostras da distribuicao geometrica.


```python
import numpy as np
import matplotlib.pyplot as plt

p=0.3
N=100000
av=np.array([])
for j in range(N):
    count=0
    x=np.random.uniform(0,1)
    while x>p:
        count=count+1
        x=np.random.uniform(0,1)
    av=np.append(av,count)

print(av)

valores=np.arange(0,16)
freq=np.array([])
for valor in valores:
    count=0
    for geomvalue in av:
        if geomvalue==valor:
            count=count+1
    freq=np.append(freq,count/N)

plt.bar(valores, freq)
plt.show()
```

### b)

Qual a probabilidade de o primeiro sucesso ocorrer na 4 tentativa?


```python
p=0.3
N=100000
value=3
av=np.array([])
count=0
for j in range(N):
    n=0
    x=np.random.uniform(0,1)
    while x>p:
        n=n+1
        x=np.random.uniform(0,1)
    av=np.append(av,n)

print(av)

for geomvalue in av:
    if geomvalue==value:
        count=count+1
prob=count/N
print("a probabilidade de o primeiro sucesso ocorrer na 4 tentativa e",prob)
```

### c)

Qual a probabilidade de precisar de mais de 6 tentativas?


```python
p=0.3
N=100000
value=5
av=np.array([])
count=0
for j in range(N):
    n=0
    x=np.random.uniform(0,1)
    while x>p:
        n=n+1
        x=np.random.uniform(0,1)
    av=np.append(av,n)

print(av)

for geomvalue in av:
    if geomvalue>value:
        count=count+1
prob=count/N
print("a probabilidade de precisar de mais de 6 tentativas e",prob)
```

### d)

Determine a media e variancia da variavel aleatoria.


```python
p=0.3
N=100000
av=np.array([])
for j in range(N):
    n=0
    x=np.random.uniform(0,1)
    while x>p:
        n=n+1
        x=np.random.uniform(0,1)
    av=np.append(av,n)

print("a media teorica e",(1-p)/p)
print("a variancia teorica e",(1-p)/(p*p))
print("a media amostral e",np.mean(av))
print("a variancia amostral e",np.var(av))
```
