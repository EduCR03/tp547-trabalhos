### c)

Com uso da transformacao inversa, desenvolva um algoritmo para gerar amostras dessa distribuicao.


```python
import numpy as np

N=10
u=np.random.uniform(0,1,N)
av=np.array([])
for iu in u:
    x=np.sqrt(iu)
    av=np.append(av,x)

print("valores uniformes")
print(u)
print("amostras geradas pela transformacao inversa")
print(av)
```

### d)

Gere 1000 amostras e compare o histograma com a pdf teorica.


```python
import matplotlib.pyplot as plt

N=1000
u=np.random.uniform(0,1,N)
Xinversa=np.sqrt(u)
print(Xinversa)

X=np.arange(0, 1, 0.01)
fx=2*X
plt.plot(X,fx)
plt.hist(Xinversa,bins=30,density=True)
plt.show()
```
