### a)

Estime `I` usando Monte Carlo com amostragem uniforme.


```python
import numpy as np
import matplotlib.pyplot as plt

N=100000
x=np.random.uniform(0,1,N)
integral=sum(np.log(1+x))/N
print("o valor estimado da integral e",integral)

X = np.arange(0, 1, 0.01)
fx = np.log(1+X)
plt.plot(X,fx)
plt.fill_between(X,fx,color='lightblue')
plt.text(0.35,0.2,f'Area = {integral:.4f}',fontsize=12)
plt.xlabel('x')
plt.ylabel('f(x)')
plt.show()
```

### b)

Execute simulacoes com `N = 10^2, 10^3, 10^4, 10^5`.


```python
Nvalue=np.array([10**2,10**3,10**4,10**5])
integralN=np.array([])

for n in Nvalue:
    x=np.random.uniform(0,1,n)
    integral=sum(np.log(1+x))/n
    integralN=np.append(integralN,integral)

print("valores de N")
print(Nvalue)
print("estimativas da integral")
print(integralN)
```

### c)

Plote o erro absoluto em funcao de `N`.


```python
Iteorico=2*np.log(2)-1
erro=np.abs(integralN-Iteorico)
print("valor teorico da integral")
print(Iteorico)
print("erro absoluto")
print(erro)

plt.plot(Nvalue,erro,'o-')
plt.xscale('log')
plt.xlabel('N')
plt.ylabel('erro absoluto')
plt.show()
```
