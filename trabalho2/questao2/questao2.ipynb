Considere a integral `I = \int_0^1 1/(1+x)^2 dx`.

### solucao analitica

Resolver a integral de forma analitica.


```python
import numpy as np
import matplotlib.pyplot as plt

Ianalitico=1/2
print("o valor analitico da integral e",Ianalitico)
```

### a)

Estime via Monte Carlo simples, usando a uniforme.


```python
N=100000
x=np.random.uniform(0,1,N)
integralsimples=sum(1/(1+x)**2)/N
errosimples=np.abs(integralsimples-Ianalitico)
print("a integral por monte carlo simples e",integralsimples)
print("o erro absoluto e",errosimples)

X=np.arange(0,1,0.01)
fx=1/(1+X)**2
plt.plot(X,fx)
plt.fill_between(X,fx,color='lightblue')
plt.text(0.35,0.6,f'Area = {integralsimples:.4f}',fontsize=12)
plt.xlabel('x')
plt.ylabel('f(x)')
plt.show()
```

### b)

Escolha uma distribuicao `p(x)`.


```python
X=np.arange(0,1,0.01)
px=2/(1+X)**2
print("a distribuicao escolhida foi p(x)=2/(1+x)**2 para 0<=x<=1")

plt.plot(X,1/(1+X)**2)
plt.plot(X,px)
plt.xlabel('x')
plt.ylabel('valor')
plt.show()
```

### c)

Reescreva o estimador e calcule a integral.


```python
N=100000
u=np.random.uniform(0,1,N)
ximp=u/(2-u)
fx=1/(1+ximp)**2
px=2/(1+ximp)**2
peso=fx/px
integralimp=sum(peso)/N
erroimp=np.abs(integralimp-Ianalitico)
print("o estimador reescrito e f(x)/p(x)")
print("o valor da integral por importance sampling e",integralimp)
print("o erro absoluto e",erroimp)
```

### d)

Compare os metodos.


```python
varsimples=np.var(1/(1+x)**2)
varimp=np.var(peso)
print("integral por monte carlo simples",integralsimples)
print("integral por importance sampling",integralimp)
print("erro do metodo simples",errosimples)
print("erro do importance sampling",erroimp)
print("variancia do metodo simples",varsimples)
print("variancia do importance sampling",varimp)

metodo=np.array([1,2])
erro=np.array([errosimples,erroimp])
plt.bar(metodo,erro)
plt.xticks([1,2],["simples","importance"])
plt.ylabel('erro absoluto')
plt.show()
```

Justifique a escolha de `p(x)`.
