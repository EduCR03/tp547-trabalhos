Deseja-se gerar amostras de `f(x)=3x^2`, para `0<=x<=1`.

### a)

Implemente aceitacao-rejeicao.

```python
import numpy as np
import matplotlib.pyplot as plt

n=100000
xfunction=np.array([])
tentativas=0

for i in range(n):
    x1=np.random.uniform(0,1)
    x2=np.random.uniform(0,1)
    tentativas=tentativas+1

    while (x2 >=x1**2):
        x1=np.random.uniform(0,1)
        x2=np.random.uniform(0,1)
        tentativas=tentativas+1

    xfunction=np.append(xfunction,x1)

print(xfunction)

X=np.arange(0,1,0.01)
plt.hist(xfunction,density=True,bins=20)
plt.plot(X,3*X**2)
plt.xlabel('x')
plt.ylabel('densidade')
plt.show()
```

### b)

Calcule taxa de aceitacao.

```python
taxaaceitacao=n/tentativas
print("a taxa de aceitacao simulada e",taxaaceitacao)
```

### c)

Compare com metodo da inversa.

```python
N=100000
u=np.random.uniform(0,1,N)
Xinversa=np.array([])

for iu in u:
    x=iu**(1/3)
    Xinversa=np.append(Xinversa,x)

print(Xinversa)
print("a media por aceitacao-rejeicao e",np.mean(xfunction))
print("a media pelo metodo da inversa e",np.mean(Xinversa))

plt.hist(Xinversa,density=True,bins=20)
plt.plot(X,3*X**2)
plt.xlabel('x')
plt.ylabel('densidade')
plt.show()
```

### d)

Discuta eficiencia.

```python
print("o metodo da inversa gera uma amostra a cada numero uniforme gerado")
print("o metodo de aceitacao-rejeicao precisou de",tentativas,"tentativas para gerar",n,"amostras")
print("por isso a inversa e mais eficiente para essa distribuicao")
```
