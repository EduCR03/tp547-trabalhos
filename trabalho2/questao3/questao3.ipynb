Deseja-se gerar amostras de `f(x)=3x^2`, para `0<=x<=1`.

### solucao analitica

Resolver a distribuicao de forma analitica.


```python
import numpy as np
import matplotlib.pyplot as plt

print("a integral de 3*x**2 de 0 ate 1 e",1)
print("a CDF e F(x)=x**3")
print("a inversa da CDF e F^-1(u)=u**(1/3)")
print("a taxa teorica de aceitacao com g(x)=1 e M=3 e",1/3)
```

### a)

Implemente aceitacao-rejeicao.


```python
n=100000
xfunction=np.array([])
tentativas=0

for i in range(n):
    x1=np.random.uniform(0,1)
    x2=np.random.uniform(0,1)
    tentativas=tentativas+1

    while (x2 >=x1**2):
        x1=np.random.uniform(0,1)
        x2=np.random.uniform(0,1)
        tentativas=tentativas+1
    xfunction = np.append(xfunction, x1)

print(xfunction)
X=np.arange(0, 1, 0.01)
fx=3*X**2
plt.plot(X,fx)
plt.hist(xfunction,bins=100,density=True)
plt.show()
```

### b)

Calcule taxa de aceitacao.


```python
taxaaceitacao=n/tentativas
print("a taxa de aceitacao simulada e",taxaaceitacao)
print("a taxa de aceitacao teorica e",1/3)
```

### c)

Compare com metodo da inversa.


```python
N=100000
u=np.random.uniform(0,1,N)
Xinversa=np.array([])

for iu in u:
    x=iu**(1/3)
    Xinversa=np.append(Xinversa,x)

print(Xinversa)
print("a media por aceitacao-rejeicao e",np.mean(xfunction))
print("a media pelo metodo da inversa e",np.mean(Xinversa))
print("a media teorica e",3/4)

plt.plot(X,fx)
plt.hist(xfunction,bins=50,density=True,alpha=0.5)
plt.hist(Xinversa,bins=50,density=True,alpha=0.5)
plt.show()
```

### d)

Discuta eficiencia.


```python
print("o metodo da inversa gera uma amostra a cada numero uniforme gerado")
print("o metodo de aceitacao-rejeicao precisou de",tentativas,"tentativas para gerar",n,"amostras")
print("por isso a inversa e mais eficiente para essa distribuicao")
print("a aceitacao-rejeicao funciona bem, mas descarta parte das amostras")
```
