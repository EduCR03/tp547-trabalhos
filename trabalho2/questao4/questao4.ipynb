## Questao 4 - Cartas

Uma urna possui 48 cartas, igualmente divididas em 4 tipos `A`, `B`, `C` e `D`, com 12 cartas de cada tipo. Retiram-se 7 cartas.

- `a)` Calcule a probabilidade de obter pelo menos uma carta `B` e uma carta `D`, sem reposicao.
- `b)` Refaça supondo reposicao.
- `c)` Compare os resultados.

```python
!pip install numpy
!pip install matplotlib
```

### a)

Sem reposicao.

```python
import numpy as np
import matplotlib.pyplot as plt

cartas=np.array(['A']*12+['B']*12+['C']*12+['D']*12)
N=100000
contasem=0

for i in range(N):
    amostra=np.random.choice(cartas,7,replace=False)
    temB=0
    temD=0

    for carta in amostra:
        if carta=='B':
            temB=1
        if carta=='D':
            temD=1

    if temB==1 and temD==1:
        contasem=contasem+1

probsim_sem=contasem/N

print("probabilidade simulada sem reposicao =",probsim_sem)
```

### b)

Com reposicao.

```python
tipos=np.array(['A','B','C','D'])
N=100000
contacom=0

for i in range(N):
    amostra=np.random.choice(tipos,7,replace=True)
    temB=0
    temD=0

    for carta in amostra:
        if carta=='B':
            temB=1
        if carta=='D':
            temD=1

    if temB==1 and temD==1:
        contacom=contacom+1

probsim_com=contacom/N

print("probabilidade simulada com reposicao =",probsim_com)
```

### c)

Comparacao entre os resultados.

```python
print("sem reposicao =",probsim_sem)
print("com reposicao =",probsim_com)
print("sem reposicao a probabilidade ficou maior")
print("com reposicao a mesma letra pode sair varias vezes")

x=np.array([1,2])
y=np.array([probsim_sem,probsim_com])

plt.bar(x,y)
plt.xticks([1,2],["sim sem","sim com"])
plt.ylabel('probabilidade')
plt.show()
```
