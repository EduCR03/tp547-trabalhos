## Questao 5 - Simulacao Monte Carlo

Considere o lancamento simultaneo de dois dados equilibrados.

- `a)` Utilize simulacao de Monte Carlo para estimar a probabilidade de que a soma dos resultados seja estritamente maior que `9`.
- `b)` Compare o valor obtido com o valor teorico.
- `c)` Discuta como o numero de simulacoes afeta a precisao do resultado.

### a)

Estimativa por Monte Carlo.

```python
import numpy as np
import matplotlib.pyplot as plt

N=100000
dado1=np.random.randint(1,7,N)
dado2=np.random.randint(1,7,N)
soma=dado1+dado2
probsim=sum(soma>9)/N

print("probabilidade simulada =",probsim)

valores=np.arange(2,13,1)
freq=np.array([])

for v in valores:
    freq=np.append(freq,sum(soma==v)/N)

plt.bar(valores,freq)
plt.xlabel('soma')
plt.ylabel('frequencia relativa')
plt.show()
```

### b)

Comparacao com o valor teorico.

```python
probteorica=6/36
erro=np.abs(probsim-probteorica)

print("probabilidade simulada =",probsim)
print("probabilidade teorica =",probteorica)
print("erro absoluto =",erro)
```

### c)

Efeito do numero de simulacoes na precisao.

```python
Nvalue=np.array([10**2,10**3,10**4,10**5])
probN=np.array([])

for n in Nvalue:
    dado1=np.random.randint(1,7,n)
    dado2=np.random.randint(1,7,n)
    soma=dado1+dado2
    prob=sum(soma>9)/n
    probN=np.append(probN,prob)

erroN=np.abs(probN-probteorica)

print("numero de simulacoes")
print(Nvalue)
print("probabilidades estimadas")
print(probN)
print("erro absoluto")
print(erroN)
print("quando N aumenta a estimativa fica mais estavel")
print("em geral o erro tende a diminuir com mais simulacoes")

plt.plot(Nvalue,erroN,'o-')
plt.xscale('log')
plt.xlabel('N')
plt.ylabel('erro absoluto')
plt.show()
```
